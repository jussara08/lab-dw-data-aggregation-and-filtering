2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
# avg monthly  premium and customer lifetime value by policy type and gender for customers who have a response "Yes" to the last marketing campaign
import pandas as pd
# Sample DataFrame
data = {
    'customer_id': [1, 2, 3, 4],
    'policy_type': ['A', 'B', 'A', 'B'],
    'gender': ['Male', 'Female', 'Female', 'Male'],
    'monthly_premium': [100, 200, 150, 250],
    'customer_lifetime_value': [1000, 2000, 1500, 2500],
    'last_campaign_response': ['Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)

# Filter the DataFrame for customers who responded "Yes" to the last campaign
filtered_df = df[df['last_campaign_response'] == 'Yes']

# Calculate the average monthly premium and customer lifetime value by policy type and gender
result = filtered_df.groupby(['policy_type', 'gender']).agg(
    avg_monthly_premium=('monthly_premium', 'mean'),
    avg_customer_lifetime_value=('customer_lifetime_value', 'mean')
).reset_index()

print(result)
#compare the insights to claim amount
import pandas as pd
# Sample DataFrame
data = {
    'customer_id': [1, 2, 3, 4],
    'policy_type': ['A', 'B', 'A', 'B'],
    'gender': ['Male', '    Female', 'Female', 'Male'],
    'monthly_premium': [100, 200, 150, 250],
    'customer_lifetime_value': [1000, 2000, 1500, 2500],
    'total_claim_amount': [500, 1500, 800, 1200],
    'last_campaign_response': ['Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)     
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')

state_counts = df.groupby("State").size().reset_index(name="Customer_Count")
filtered_states = state_counts[state_counts["Customer_Count"] > 500]

print(filtered_states)
```
